```python
#we want our x to be test, y to be speed, hue to be graphics, so we must reformat dataframe in that way
import seaborn as sns
import matplotlib.pyplot as plt
df["Test"] = df.index
df["Test name"] = ["MobileNet-V2", "MobileNet-V2", "Inception-V3", "Inception-V3", "Inception-V4", "Inception-V4", "Inc-ResNet-V2", "Inc-ResNet-V2", "ResNet-V2-50", "ResNet-V2-50", "ResNet-V2-152", "ResNet-V2-152", "VGG-16", "VGG-16", "SRCNN 9-5-5", "SRCNN 9-5-5", "SRCNN 9-5-5", "VGG-19 Super-Res", "VGG-19 Super-Res", "VGG-19 Super-Res", "ResNet-SRGAN", "ResNet-SRGAN", "ResNet-SRGAN", "ResNet-DPED", "ResNet-DPED", "ResNet-DPED", "U-Net", "U-Net", "U-Net", "Nvidia-SPADE", "Nvidia-SPADE", "ICNet", "ICNet", "PSPNet", "PSPNet", "DeepLab", "DeepLab", "Pixel-RNN", "Pixel-RNN", "LSTM", "LSTM", "GNMT"]

new_df = pd.melt(df[:20], id_vars=["Test", "Test name"], value_vars=['i7-6700HQ Integrated GPU', 'GTX 960M', 'GTX 1080']).rename(columns={"value":"Time (Milliseconds)"})
#print(new_df)
fig, ax = plt.subplots(figsize=(8,5))
ax = sns.catplot(x="Test", y="Time (Milliseconds)", hue="variable",kind="bar", data=new_df, ax=ax)
plt.close(2)
plt.show()
```
